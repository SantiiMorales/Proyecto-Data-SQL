# Visual Data

Aqui lo que haremos sera abrir los archivos csv, y hacer una grafica basica de cada uno de ellos

```python
#Importacion de librerias
import pandas as pd 
import numpy as np
from matplotlib import pyplot as plt 
```

```python
#Abrimos los archivos y los transformamos en dataframes
df_productos = pd.read_csv('Producto_mas_vendido.csv')
df_total_mes = pd.read_csv('total_mes.csv')
df_ventas_cliente = pd.read_csv('Total_ventas_cliente.csv')
```

### Creacion de funcion para hacer el grafico

```python
#Ya que los datos recopilados son muy basicos haremos un grafico igual de sencillo
def grafico_bar(dfi, figname, Eje_X, Eje_Y, Title_X, Title_Y, tinta):
    df = dfi.copy()
    df = df[[Eje_X, Eje_Y]].groupby(Eje_X).sum()
    df = df.reset_index()
    df = df.sort_values(by=Eje_Y, ascending=True)
    
    plt.bar(df[Eje_X], df[Eje_Y], width=0.3, color=tinta)
    plt.title(figname)
    plt.xlabel(Title_X)
    plt.ylabel(Title_Y)   
    plt.xticks(df[Eje_X], rotation=45)

    plt.show()
    
    return
grafico_bar(df_productos, 'Productos Mas Vendidos', 'nombre', 'cantidad_vendida','Nombre del Producto', 'Cantidad', 'Red')
```

```python
grafico_bar(df_ventas_cliente, 'Ventas por cliente', 'Nombre', 'total_ventas','Nombre del Cliente', 'Total de ventas', 'Darkblue')
```
